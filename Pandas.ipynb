## Indexing
Both loc and iloc are row-first, column second types

### Index Based Selection
**iloc**: selecting data based on its numberical position in data.

1. Like Python end index is excluded


```python
wine_reviews['country'][0]
```

```python
wine_reviews.iloc[:1,]
```

```python
wine_reviews.iloc[:,:5]
```

```python
wine_reviews.iloc[1:4,2:5]
```

```python
wine_reviews.iloc[[1, 2, 7,100], ]
```

```python
wine_reviews.iloc[[1, 2, 7,100],[0,4,5,10,12] ]
```

```python
wine_reviews.iloc[-5:]
```

```python
wine_reviews.iloc[-5:,[0,4,5]]
```

```python
wine_reviews.iloc[-5:,[-2,8,-1]]
```

### Label Based Selection

**loc**: data index values matters not its position 

1. Last index included unlike python
2. we want to select "all the alphabetical fruit choices between Apples and Potatoes", then it's a lot more convenient to index df.loc['Apples':'Potatoes']

```python
wine_reviews.loc[0,'country']
```

```python
wine_reviews.loc[:,'country']
```

```python
wine_reviews.loc[:5,['country','price','variety','winery']]
```

```python
wine_reviews.loc[:5,['country','price','variety','winery']]
```

### Maps

It takes one set of values and transformed into another set of values

```python
wine_reviews_points_mean = wine_reviews.points.mean()
wine_reviews.points.map(lambda x:x-wine_reviews_points_mean)
```

```python
def remean_points(row):
    row.points-=wine_reviews_points_mean
    return row
wine_reviews.apply(remean_points, axis=1)
```

```python
wine_reviews.apply(remean_points, axis='columns')
```

```python
wine_reviews.points-wine_reviews.points.mean()
```
